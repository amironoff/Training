# Indexing & slicing data

```python
# Get single feature vector
df['churn'].mean()
```

```python
# Now, some logical filtering, using boolean predicates
df[df['churn'] == 1].mean()
```

```python
# Let's combine stuff above to answer this question:
# On average, how many minutes per day do churned users speak?

df[df["churn"] == 1]["total day minutes"].mean()
```

```python
# What's the max international call length for loyal users that don't use roaming
df[(df['churn'] == 0) & (df['international plan'] == 'no')]['total intl minutes'].max()
```

```python
# Датафреймы можно индексировать как по названию столбца или строки,
# так и по порядковому номеру.
# Для индексации по названию используется метод loc, по номеру — iloc.
```

```python
# передай нам значения первых пяти строк в столбцах от State до Area code

df.loc[0:5, 'state':'area code']
```

```python
# передай нам значения первых пяти строк в первых трёх столбцах

df.iloc[0:5, 0:3]
```

# Transforming the dataset with functions

```python
# Применение функции к каждому столбцу: apply

# Метод apply можно использовать и для того,
# чтобы применить функцию к каждой строке. 
# Для этого нужно указать axis=1.

df.apply(np.max)
```

```python
# Применение функции к каждой ячейке столбца: map

# Например, метод map можно использовать
# для замены значений в колонке,
# передав ему в качестве аргумента 
# словарь вида {old_value: new_value}:

d = {'no': False, 'yes': True}
df['international plan'] = df['international plan'].map(d)
df.head()
```

# Grouping data

```python
# В общем случае группировка данных в Pandas выглядит следующим образом:
# df.groupby(by=grouping_columns)[columns_to_show].function()
```

```python
columns_to_show = ['total day minutes', 'total eve minutes', 'total night minutes']

df.groupby(['churn'])[columns_to_show].describe(percentiles=[])
```

```python
df.groupby(['churn'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

# Analyzing feature dependencies with crosstab

```python
pd.crosstab(df['churn'], df['international plan'])
```
